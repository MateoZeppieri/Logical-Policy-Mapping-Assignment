## Section 3 — Classify Logical Constraint Types

1. **Rule:** If the star hitter is signed, then total payroll must remain under the budget limit.

   * **Type:** Activation
   * **Explanation:** Signing the star hitter triggers the requirement to satisfy the payroll constraint, so choosing A (star hitter) requires B (budget compliance).
   * **Ambiguity:** None—this is a standard activation constraint tied to a resource limit.

2. **Rule:** If a veteran pitcher is signed, then a prospect cannot be promoted.

   * **Type:** Suppression
   * **Explanation:** Choosing the veteran pitcher prevents the selection of the prospect, which matches a suppression relationship.
   * **Ambiguity:** Could clarify whether this applies only to the same position or all prospects.

3. **Rule:** If an extra pitcher is carried, then the number of hitters selected must decrease.

   * **Type:** Activation
   * **Explanation:** Carrying an extra pitcher forces an adjustment elsewhere (fewer hitters), so selecting A requires a related outcome B.
   * **Ambiguity:** Needs clarification on exactly how many hitters must be reduced.

4. **Rule:** At least one of Player A or Player B must be signed.

   * **Type:** Either-or
   * **Explanation:** The rule requires at least one of two options to be chosen, which directly matches the either-or definition.
   * **Ambiguity:** Could clarify whether signing both is allowed or if exactly one is preferred.

5. **Rule:** If Player B is signed, then Player C cannot be signed.

   * **Type:** Mutually exclusive
   * **Explanation:** The rule prevents both players from being selected together, meaning at most one can be chosen.
   * **Ambiguity:** Could confirm if the restriction works both ways (B excludes C and C excludes B).

6. **Rule:** If a prospect is promoted, then a veteran pitcher must also be signed.

   * **Type:** Activation
   * **Explanation:** Promoting a prospect requires also selecting a veteran, so A requires B.
   * **Ambiguity:** Could clarify if the reverse is required (i.e., whether they must always be chosen together).

7. **Rule:** The team cannot sign Player A, Player B, and Player C all at the same time.

   * **Type:** Suppression
   * **Explanation:** This rule prevents a specific combination of decisions, which is a form of suppression of a joint outcome.
   * **Ambiguity:** Could clarify whether any two are allowed or if there are additional limits.

## Section 5 — Feasibility, Conflicts, and "No Feasible Solution"

A potential conflict arises between Policy 2, Policy 6, and the general need to fill roster spots. Policy 2 states that if a veteran pitcher is signed, then a prospect cannot be promoted, while Policy 6 states that if a prospect is promoted, then a veteran pitcher must also be signed. Together, these create a circular dependency where promoting a prospect requires a veteran, but signing a veteran prevents promoting the prospect. This conflict can make it impossible to satisfy both policies at the same time.

This could make the model infeasible because there may be situations where the team needs to promote a prospect due to budget constraints or lack of available players, but the policies prevent any valid combination of decisions. A "no feasible solution" scenario would look like a general manager trying to finalize the roster but finding that every possible combination of players violates at least one rule. For example, if the team cannot afford enough veteran players but is also not allowed to promote prospects without them, then no valid roster can be formed.

If this happens, the first policy to reconsider would be Policy 2 or Policy 6. These two rules directly conflict, so one of them likely needs to be relaxed or clarified. For example, the team might allow both a veteran and a prospect to be selected together or remove the restriction that prevents prospects from being promoted when a veteran is signed. These adjustments would restore feasibility while still maintaining the intent of the policies.

To explain this to a non technical stakeholder, I would say that the current rules contradict each other and leave no possible way to build a valid roster. I would explain that some policies are too strict when combined and that we need to slightly adjust them to allow a workable solution.
